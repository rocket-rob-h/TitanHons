```python

# -*- coding: utf-8 -*-
"""
The function below performs stagnation point radiative heat flux
calculations for Earth entry, based on Tauber and Sutton (1991):
Tauber, M. E., & Sutton, K. (1991).
Stagnation-point radiative heating relations for Earth and Mars entries.
Journal of Spacecraft and Rockets, 28(1), 40-42.
"""
VERSION_STRING = '07-Apr-2021'
def tauber_and_sutton_earth_stag_point_radiative_heat_flux(rho_inf, U_inf, R_n):
    """
    This function performs stagnation point radiative heat flux
    calculations for Earth entry, based on Tauber and Sutton (1991):
    Tauber, M. E., & Sutton, K. (1991).
    Stagnation-point radiative heating relations for Earth and Mars entries.
    Journal of Spacecraft and Rockets, 28(1), 40-42.
    Inputs are in si units, with an output in W/cm**2, sadly...
    :param rho_inf: freestream density in kg/m**3
    :param U_inf: freestream velocity in m/s
    :param R_n: nose radius in m
    :return: q_r in W/cm**2
    """
import numpy as np
C = 4.736e4
# this comes from the text, isn't so obvious...
if R_n < 1.0:
    a_max = 1.0
elif 1.0 <= R_n <= 2.0:
    a_max = 0.6
elif 2.0 <= R_n <= 3.0:
    a_max = 0.5
#print "a_max = {0}".format(a_max)
a_calculated = 1.072e6 * U_inf ** -1.88 * rho_inf** -0.325
#print "a_calculated = {0}".format(a_calculated)
a = min([a_max, a_calculated])
#print "a = {0}".format(a)
b = 1.22
# now we need the f_U value, which they said can be found through
# linear interpolation...
# so we'll enter both lists and add some interpolation
U_list = [9000.0, 9250.0, 9500.0, 9750.0, 10000.0, 10250.0,10500.0, 10750.0, 11000.0,
11500.0, 12000.0, 12500.0, 13000.0, 13500.0, 14000.0, 14500.0, 15000.0,
15500.0, 16000.0]
f_U_list = [1.5, 4.3, 9.7, 19.5, 35.0, 55.0, 81.0, 115.0, 151.0, 238.0, 359.0,
495.0, 660.0, 850.0, 1065.0, 1313.0, 1550.0, 1780.0, 2040.0]
f_U = np.interp(U_inf, U_list, f_U_list)
q_r = C * (R_n ** a) * (rho_inf ** b) * f_U # W/cm**2
return q_r
```
